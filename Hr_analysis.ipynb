## Data preprocessing

```python
# dropping unwanted columns
df.drop(['EmployeeCount','id'], axis =1, inplace=True)
```

```python
attrition_dummies = pd.get_dummies(df['Attrition'])
attrition_dummies.head()
```

```python
df = pd.concat([df, attrition_dummies], axis =1)
df.head()
```

```python
df.drop(['Attrition',0], axis = 1, inplace=True)
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(df.corr(), annot=True, fmt = '.0%')
```

```python
df.drop(['Age','JobLevel'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
    if df[column].dtype==np.number:
        continue
        
    else:
        df[column]=LabelEncoder().fit_transform(df[column])
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 10, criterion ='entropy', random_state=0)
```

```python
X = df.drop([1], axis=1)
y = df[1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
### Prediction for test data
pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(X_train, y_train)
sv.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = sv.predict(X_train)
confusion_matrix(y_train,y_pred)
```
